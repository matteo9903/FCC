```python
from transformers import AutoTokenizer

# Load a pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")

# Example text
text = "I was so not happy with the Barbie Movie"

# Tokenize the text
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)

# Convert tokens to input IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)

# Encode the text (tokenization + converting to input IDs)
encoded_input = tokenizer(text)
print("Encoded Input:", encoded_input)

# Decode the text
decoded_output = tokenizer.decode(input_ids)
print("Decode Output: ", decoded_output)
```

**token_type_ids**<br>
These IDs are used to distinguish between different sequences in tasks that involve multiple sentences, such as question-answering and sentence-pair classification. BERT uses this mechanism to understand which tokens belong to which segment. For single-sequence tasks like sentiment analysis, token_type_ids are all zeros.

**attention_mask** <br>
The attention mask is used to differentiate between actual tokens and padding tokens (if any). It helps the model focus on non-padding tokens and ignore padding tokens. A value of 1 indicates that the token should be attended to, while a value of 0 indicates padding.

**Why Padding Tokens Are Used**<br>
Uniform Sequence Length: Deep learning models typically process input data in batches. To efficiently process these batches, all sequences in a batch must have the same length. Padding tokens ensure this by extending shorter sequences to match the length of the longest sequence in the batch.
Efficient Computation: Fixed-length sequences allow for more efficient use of hardware resources, as the model can process all sequences in parallel without needing to handle variable-length sequences individually.

## Step 2: Load and Prepare the Dataset

```python
# carico dataset direttamente da Hugging Face semplicemente usando la funzione
# load_dataset(nome_dataset)
from datasets import load_dataset
dataset = load_dataset('imdb')
```

```python
dataset
# uso solo training e test set
```

```python
dataset["train"][0]
```

## Step 3: Preprocess the Data
Tokenize the dataset using the tokenizer associated with the pre-trained model.

**È importantissimo che modello per il tokenizer sia lo stesso che quello del training**

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the dataset
def tokenize_function(examples):
  # seleziono frase con massima lunghezza e ad altre applico padding mettendo 0,
  # questo per facilitare training avendo sempre stessa dimensione
    return tokenizer(examples['text'], padding="max_length", truncation=True)

# tokenizzo i dataset usando funzione "map" --> specifica di Hugging Face
tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets  # --> vedo che input IDs e tokens
```

```python
tokenized_datasets["train"][0]
```

con questi step ho effettuato conversione testo nei vettori adatti ad essere usati per il training del modello

## Step 4: Set Up the Training Arguments
Specify the hyperparameters and training settings.

**basta seguire codice HF, non serve ricordare tutto a memoria!!**

```python
from transformers import TrainingArguments
# posso specificare i parametri del modello
training_args = TrainingArguments(
    output_dir='./results',          # Output directory --> per salvare modello
    eval_strategy ="epoch",     # Evaluate every epoch
    learning_rate=2e-5,              # Learning rate
    per_device_train_batch_size=8,  # Batch size for training
    per_device_eval_batch_size=8,   # Batch size for evaluation
    num_train_epochs=1,              # Number of training epochs
    weight_decay=0.01,               # Strength of weight decay
    gradient_accumulation_steps=2  # accumula gradiente per steps, simulando
                                   # un batch size più grande (in questo caso 16)
)
training_args
```

## Step 5: Initialize the Model
Load the pre-trained model and define the training procedure.

```python
from transformers import AutoModelForSequenceClassification, Trainer

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased',
                                                           num_labels=2)
      # num_labels --> essendo sentiment analysis voglio solo 2 label;
      #                positivo/negativo

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)
# non ho ancora fatto il training però, devo utilizzare un comando apposito;
# per ora ho solo inizializzato i parametri e weights della rete
```

## Step 6: Train the Model
Fine-tune the pre-trained model on your specific dataset.

```python
# Train the model

# comando include login e funzionalità di WANDB
# esso consente
trainer.train()

# alla fine vedo training loss e validation loss
# incrementando numero di epoche, vedrò anche diminuire loss
```

## Step 7: Evaluate the Model
Assess the model's performance on a validation set.

```python
# Evaluate the model
results = trainer.evaluate()
print(results)

# mi focalizzo su loss, più è piccolo e vicino a 0 e meglio è

# ho anche tempo messo per la valutazione + altri parametri
```

## Step 8: Save the Fine-Tuned Model
Save the fine-tuned model for later use.

* salvo modello in uno specifico percorso per poterlo eventualmente riutilizzare tramite i **checkpoint**
* importante è anche **salvare il tokenizer** perché esso è stato **fittato sul training set del modello di base,** per cui se si vuole utilizzare modello con nuovi dati e importante mantenere anche il tokenizer per avere una certa *consistenza dei passaggi*

```python
# Save the model
model.save_pretrained('./fine-tuned-model')
tokenizer.save_pretrained('./fine-tuned-tokenizer')
```

```python
# importo librerie necessarie
import arxiv
import pandas as pd
```

```python
# Query to fetch AI-related papers --> estraggo paper che riguardano AI e ML
# limito ai primi 10 risultati
query = 'ai OR artificial intelligence OR machine learning'
search = arxiv.Search(query=query, max_results=10, sort_by=arxiv.SortCriterion.SubmittedDate)

# Fetch papers
papers = []
for result in search.results():
  # estraggo info su data pubblicazione, titolo, abstract e categoria
    papers.append({
      'published': result.published,
        'title': result.title,
        'abstract': result.summary,
        'categories': result.categories
    })

# Convert to DataFrame
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
# Example abstract from API
abstract = df['abstract'][0]

# scarico modello per sommarizzare l'articolo
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# Summarization --> eseguo task su uno specifico abstract
summarization_result = summarizer(abstract)
```
